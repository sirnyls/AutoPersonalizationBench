```python
import pandas as pd
import json
import random

with open('datasets/TruthfulQA.jsonl', 'r') as file:
    lines = file.readlines()
truthfulQA_json = [json.loads(line) for line in lines]

with open('datasets/NQ1.jsonl', 'r') as file:
    lines = file.readlines()
NQ1_json = [json.loads(line) for line in lines]

with open('datasets/NQ2.jsonl', 'r') as file:
    lines = file.readlines()
NQ2_json = [json.loads(line) for line in lines]

with open('datasets/Boolq.jsonl', 'r') as file:
    lines = file.readlines()
Boolq_json = [json.loads(line) for line in lines]
```

```python
import os
from openai import OpenAI

client = OpenAI(
    api_key=os.environ.get("OPENAI_API_KEY"),
)
def generate_statements (json_object):
    for obj in json_object:
        statement = obj['adv']['control']
        #print(statement)
        prompt = f"""Generate another factual wrong statement that is from the domain of the statement provided.
                    Statement: {statement}\n Your Output:"""
        response = client.chat.completions.create(
        model="gpt-4-1106-preview",
        messages=[
            {
            "role": "system",
            "content": "Your task is to generate a factual wrong statement."
            },
            {
            "role": "user",
            "content":  prompt
            }
        ]
        )
        output = response.choices[0].message.content
        #print(output)
        obj['adv']['wrong_statement_same_domain'] = output
    for obj in json_object:
        statement = obj['adv']['control']
        print(statement)
        prompt = f"""Generate another factual wrong statement that is from a foreign of the statement provided.
                    Statement: {statement}\n Your Output:"""
        response = client.chat.completions.create(
        model="gpt-4-1106-preview",
        messages=[
            {
            "role": "system",
            "content": "Your task is to generate a factual wrong statement."
            },
            {
            "role": "user",
            "content":  prompt
            }
        ]
        )
        output = response.choices[0].message.content
        print(output)
        obj['adv']['wrong_statement_foreign_domain'] = output
    for obj in json_object:
        statement = obj['adv']['wrong_statement_foreign_domain']
        print(statement)
        prompt = f"""Generate a correct statement based on a wrong statement provided.
                    Statement: {statement}\n Your Output:"""
        response = client.chat.completions.create(
        model="gpt-4-1106-preview",
        messages=[
            {
            "role": "system",
            "content": "Your task is to generate a corresponding correct statement."
            },
            {
            "role": "user",
            "content":  prompt
            }
        ]
        )
        output = response.choices[0].message.content
        print(output)
        obj['adv']['right_statement_foreign_domain'] = output
    for obj in json_object:
        statement = obj['adv']['wrong_statement_same_domain']
        print(statement)
        prompt = f"""Generate a correct statement based on a wrong statement provided.
                    Statement: {statement}\n Your Output:"""
        response = client.chat.completions.create(
        model="gpt-4-1106-preview",
        messages=[
            {
            "role": "system",
            "content": "Your task is to generate a corresponding correct statement."
            },
            {
            "role": "user",
            "content":  prompt
            }
        ]
        )
        output = response.choices[0].message.content
        print(output)
        obj['adv']['right_statement_same_domain'] = output
```

```python
generate_statements(truthfulQA_json)
```

```python
generate_statements(NQ1_json)
```

```python
generate_statements(NQ2_json)
```

```python
generate_statements(Boolq_json)
```

```python
with open('datasets/TruthfulQA.jsonl', 'w') as file:
    for obj in truthfulQA_json:
        # Convert the object to a JSON string and write it to the file
        json_string = json.dumps(obj)
        file.write(json_string + '\n')
with open('datasets/Boolq.jsonl', 'w') as file:
    for obj in Boolq_json:
        # Convert the object to a JSON string and write it to the file
        json_string = json.dumps(obj)
        file.write(json_string + '\n')
with open('datasets/NQ1.jsonl', 'w') as file:
    for obj in NQ1_json:
        # Convert the object to a JSON string and write it to the file
        json_string = json.dumps(obj)
        file.write(json_string + '\n')
with open('datasets/NQ2.jsonl', 'w') as file:
    for obj in NQ2_json:
        # Convert the object to a JSON string and write it to the file
        json_string = json.dumps(obj)
        file.write(json_string + '\n')
```
